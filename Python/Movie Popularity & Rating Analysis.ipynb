```python
df.drop(columns=["Unnamed: 0"],inplace = True)
```

```python
df["release_date"]=pd.to_datetime(df["release_date"])
df["year"]=df["release_date"].dt.year
```

```python
df[["vote_average", "vote_count"]].describe()
```

```python
df.sort_values("vote_count", ascending=False)[["title", "vote_average", "vote_count"]].head(10)
```

```python
popular_movies = df[df["vote_count"]>1000]
popular_movies.sort_values("vote_count",ascending=False)[["title", "vote_average", "vote_count"]].head(10)
```

```python
df[["vote_average", "vote_count"]].corr()
```

```python
df.groupby("year")["vote_average"].mean().sort_index()
```

```python
movie_stats = df.groupby("year").agg({
    "vote_average": "mean",
    "vote_count": "sum"
}).reset_index()

movie_stats.to_csv("movie_yearly_stats.csv", index=False)
```

```python
movie_stats["year"] = movie_stats["year"].astype(int)
movie_stats.to_csv("movie_yearly_stats_fixed.csv", index=False)
```
